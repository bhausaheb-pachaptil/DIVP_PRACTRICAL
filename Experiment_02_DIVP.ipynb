```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Upload an image
uploaded = files.upload()

image_name = list(uploaded.keys())[0]
print("Uploaded image:", image_name)

# Read image using OpenCV
file_bytes = np.frombuffer(uploaded[image_name], np.uint8)
img_bgr = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

# Convert BGR â†’ RGB for display
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Get original dimensions
h, w = img_rgb.shape[:2]
print("Original size:", w, "x", h)

# Resize: scale down to 50%, scale up to 150%
resized_small = cv2.resize(img_rgb, (0, 0), fx=0.5, fy=0.5)   # half size
resized_large = cv2.resize(img_rgb, (0, 0), fx=1.5, fy=1.5)   # 1.5 times

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(resized_small)
plt.title("Resized 50%")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(resized_large)
plt.title("Resized 150%")
plt.axis("off")

plt.show()
```

```python
# Rotate image by 45 degrees around its center
h, w = img_rgb.shape[:2]
center = (w // 2, h // 2)

angle = 45  # degrees
scale = 1.0

# Get rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)

# Apply affine transform
rotated = cv2.warpAffine(img_rgb, rotation_matrix, (w, h))

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(rotated)
plt.title("Rotated 45Â°")
plt.axis("off")

plt.show()
```

```python
# Simple crop: take center region

h, w = img_rgb.shape[:2]

# Define crop region (you can adjust these)
start_row = h // 4
end_row   = 3 * h // 4
start_col = w // 4
end_col   = 3 * w // 4

cropped = img_rgb[start_row:end_row, start_col:end_col]

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cropped)
plt.title("Cropped Center Region")
plt.axis("off")

plt.show()
```
