```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 2: Upload an image
uploaded = files.upload()

# Get file name
image_name = list(uploaded.keys())[0]
print("Uploaded image:", image_name)

# Read image in BGR format
file_bytes = np.frombuffer(uploaded[image_name], np.uint8)
img_bgr = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

# Convert to grayscale
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

# Show original grayscale image
plt.imshow(img_gray, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Step 3: Mean filter (smoothing using average)
# 3x3 kernel
mean_filtered = cv2.blur(img_gray, (3, 3))

plt.imshow(mean_filtered, cmap='gray')
plt.title("Mean Filter Output (3x3)")
plt.axis("off")
plt.show()
```

```python
# Step 4: Min filter (using erosion)
kernel = np.ones((3, 3), np.uint8)  # 3x3 structuring element
min_filtered = cv2.erode(img_gray, kernel)

plt.imshow(min_filtered, cmap='gray')
plt.title("Min Filter Output (Erosion 3x3)")
plt.axis("off")
plt.show()
```

```python
# Step 5: Max filter (using dilation)
kernel = np.ones((3, 3), np.uint8)  # 3x3 structuring element
max_filtered = cv2.dilate(img_gray, kernel)

plt.imshow(max_filtered, cmap='gray')
plt.title("Max Filter Output (Dilation 3x3)")
plt.axis("off")
plt.show()
```

```python
# Step 6: Median filter (good for salt & pepper noise)
# Kernel size must be odd: 3,5,7,...
median_filtered = cv2.medianBlur(img_gray, 3)

plt.imshow(median_filtered, cmap='gray')
plt.title("Median Filter Output (3x3)")
plt.axis("off")
plt.show()
```

```python
# Step 7: Sobel filter for edge detection

# Sobel X (horizontal edges)
sobel_x = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_x_abs = np.uint8(np.absolute(sobel_x))

plt.imshow(sobel_x_abs, cmap='gray')
plt.title("Sobel Filter Output - Gx (Horizontal Edges)")
plt.axis("off")
plt.show()

# Sobel Y (vertical edges)
sobel_y = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=3)
sobel_y_abs = np.uint8(np.absolute(sobel_y))

plt.imshow(sobel_y_abs, cmap='gray')
plt.title("Sobel Filter Output - Gy (Vertical Edges)")
plt.axis("off")
plt.show()

# Gradient magnitude
sobel_mag = np.sqrt(sobel_x**2 + sobel_y**2)
sobel_mag = np.uint8(np.absolute(sobel_mag))

plt.imshow(sobel_mag, cmap='gray')
plt.title("Sobel Gradient Magnitude Output")
plt.axis("off")
plt.show()
```

```python
# Step 8: LoG filter (Laplacian of Gaussian)

# Step 8.1: Gaussian smoothing
gaussian_blur = cv2.GaussianBlur(img_gray, (3, 3), 0)

plt.imshow(gaussian_blur, cmap='gray')
plt.title("Gaussian Smoothed Image (Before LoG)")
plt.axis("off")
plt.show()

# Step 8.2: Laplacian for edge detection
log = cv2.Laplacian(gaussian_blur, cv2.CV_64F)
log_abs = np.uint8(np.absolute(log))

plt.imshow(log_abs, cmap='gray')
plt.title("LoG Filter Output (Laplacian of Gaussian)")
plt.axis("off")
plt.show()
```
